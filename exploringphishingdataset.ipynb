```python
import re
def clean_text(text):
  if isinstance(text, str):
      text = re.sub(r'<[^>]+>', '', text) #for html
      text = re.sub(r'[^a-zA-Z]', ' ', text)  #for non-alphabetic
      text = ' '.join(text.split())  #removing extra space
      return text.lower()  #lowercase
  else:
    return str(text)

df['Text'] = df['Text'].apply(clean_text)

#Tokenization
df['Text'] = df['Text'].str.split()

#Stopword removal
stopwords = set(['this', 'is', 'to', 'your', 'and', 'the', 'in', 'an', 'a'])
df['Text'] = df['Text'].apply(lambda x: [word for word in x if word not in stopwords])
```

```python
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer

stemmer = PorterStemmer()
df['Text'] = df['Text'].apply(lambda x: [stemmer.stem(word) for word in x])

#Feature Extraction
tfidf_vectorizer = TfidfVectorizer(max_features=1000)
X_tfidf = tfidf_vectorizer.fit_transform(df['Text'].apply(lambda x: ' '.join(x)))
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X_tfidf.toarray())
```

```python
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(df['Type'])

selector = SelectKBest(chi2, k=100)
X_selected = selector.fit_transform(X_tfidf, y_encoded)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_selected, y_encoded, test_size=0.2, random_state=42)
```
